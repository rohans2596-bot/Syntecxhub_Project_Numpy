```python
import pandas as pd

# Create sample data
data = {
    'Date': ['2024-01-05','2024-01-15','2024-02-10','2024-02-20','2024-03-08','2024-03-18'],
    'Category': ['Electronics','Clothing','Electronics','Clothing','Electronics','Clothing'],
    'Sales': [1200,800,1500,700,1800,900]
}

df = pd.DataFrame(data)

# Save as CSV in Colab's file system
df.to_csv('sales_data.csv', index=False)

print("CSV file created successfully!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("sales_data.csv")
df['Date'] = pd.to_datetime(df['Date'])
```

ðŸ“ˆ Sales Over Time (Line Chart)

```python
#ðŸ“ˆ Sales Over Time (Line Chart)
daily_sales = df.groupby('Date')['Sales'].sum()

daily_sales.plot()
plt.xlabel("Date")
plt.ylabel("Sales")
plt.title("Sales Over Time")
plt.show()
```

ðŸ“† Monthly Sales

```python
monthly_sales = df.resample('M', on='Date')['Sales'].sum()

monthly_sales.plot()
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

ðŸ“Š Category Comparison (Bar Chart)

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='bar')
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.title("Sales by Category")
plt.show()
```

ðŸ¥§ Category Share (Pie Chart)

```python
category_sales.plot(kind='pie', autopct='%1.1f%%')
plt.ylabel("")
plt.title("Sales Share by Category")
plt.show()
```
